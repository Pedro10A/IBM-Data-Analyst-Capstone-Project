```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT *
FROM master
"""
df_master = pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.hist(df_master['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')

plt.tight_layout()
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

plt.figure(figsize=(8, 6))

sns.boxplot(x=df_master['Age'], color='skyblue')

plt.xlabel('Age')
plt.title('Box Plot of Age')

plt.tight_layout()
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.figure(figsize=(10, 6))

plt.scatter(df_master['Age'], df_master['WorkWeekHrs'], color='skyblue', alpha=0.6)

plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs Work Week Hours')

plt.tight_layout()
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(15,8))

plt.scatter(df_master['WorkWeekHrs'], df_master['CodeRevHrs'], s=df_master['Age']*20, alpha=0.6, c='skyblue', edgecolors='w', linewidth=1)

plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')

plt.tight_layout()
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
conn = sqlite3.connect("m4_survey_data.sqlite")

query = "SELECT Respondent, DatabaseDesireNextYear FROM DatabaseDesireNextYear;"
df_next = pd.read_sql_query(query, conn)

database_counts = df_next['DatabaseDesireNextYear'].value_counts()

top_databases = database_counts.head(5)

plt.figure(figsize=(8, 6))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.axis('equal')
plt.show()

conn.close()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_filtered = df_master[(df_master['Age'] >= 30) & (df_master['Age'] <= 35)]

medians = df_filtered.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

medians.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xticks(rotation=0)
plt.legend(loc='upper left')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_filtered = df_master[(df_master['Age'] >= 45) & (df_master['Age'] <= 60)]

medians = df_filtered.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 6))
plt.plot(medians.index, medians.values, marker='o')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
main_branch_counts = df_master['MainBranch'].value_counts()

plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branch')
plt.show()
```
